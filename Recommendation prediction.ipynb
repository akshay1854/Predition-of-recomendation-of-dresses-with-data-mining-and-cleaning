```python
clf=RandomForestClassifier()
```

```python
grid_param = {'n_estimators' : [100,200,300,400,500], 
              'criterion':['entropy','gini'],
              'bootstrap':[True,False]}
```

```python
from sklearn.model_selection import GridSearchCV
gd_sr = GridSearchCV(estimator = clf,param_grid= grid_param, scoring = 'accuracy',cv = 5, n_jobs = -1)
```

```python
gd_sr.fit(x_train,y_train)
```

```python
best_parameters = gd_sr.best_params_
print(best_parameters)
```

```python
best_result= gd_sr.best_score_
print(best_result)
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.feature_selection import RFECV
```

```python
rfecv = RFECV(estimator=clf, step=1, cv =StratifiedKFold(10),scoring='accuracy')
rfecv.fit(x_train,y_train)
```

```python
print('optimal feature : {}'.format(rfecv.n_features_))
```

```python
plt.figure(figsize = (16,9))
plt.title('REFCV', fontsize = 18, fontweight = 'bold', pad=20)
plt.xlabel('no. of features selected',fontsize = 14, labelpad = 20 )
plt.ylabel('% correct classfication',fontsize = 14, labelpad = 20 )
plt.plot(range(1, len(rfecv.grid_scores_) + 1),rfecv.grid_scores_)
plt.show()
```
